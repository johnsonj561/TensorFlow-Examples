### Deep Neural Network with Keras
- 3 layer DNN built with Keras and TensorFlow backend
- [Pima Indians Diabetes Dataset](https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.names) contains 768 instances, 200 of which are labelled positive for diabetes
    - each sample includes 8 features

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
# load data
dataset = numpy.loadtxt("data.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
# dense: 1 x 8 --> 1 x 12
model.add(Dense(12, input_dim=8, activation='relu'))
# dense: 1 x 12 --> 1 x 8
model.add(Dense(8, activation='relu'))
# dense: 1 x 8 --> 1 x 1
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
